```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
plt.plot(x,y)
#This is used to plot a graph between x and y
plt.xlabel('X axis');
plt.ylabel('Y axis');
#used to fix x and y label
plt.title('String Title');
# use to fix the title

plt.ylim(0,15);
#used to set limit for y axis
plt.xlim(0,6);
#used to set limit for x axis

#use semicolon to avoid datatype to be displayed 

plt.savefig('myfirstplot.png')
```

```python
a = np.arange(1,10)
plt.plot(a,'X-.',color='black', ms=10);
#in this the 'X:' is used to mark the points with x and join them using dottted lines
# nd the color is black
# here the ms is the marker size and is used to set the size of the marker
```

```python
a = np.linspace(0,10,11)
a
```

```python
b = a**4
```

```python
fig = plt.figure()
#Large Axis
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);

#Small Axis
axes2 = fig.add_axes([1,1,2,2])
axes2.plot(a,b);

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('a')
axes1.set_ylabel('a power 4')
axes1.set_title('Power')


axes2.set_xlim(1,2)
axes2.set_ylim(0,30)
axes2.set_xlabel('a')
axes2.set_ylabel('a power 4')
axes2.set_title('Zoomed In');
```
